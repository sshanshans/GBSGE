```python
def geometric_series_sum(z, K):
    """
    Computes the sum of the geometric series: sum_{k=1}^K z^k for |z| < 1.

    Parameters:
    z (float or complex): The common ratio of the series, |z| < 1.
    K (int): The number of terms in the series.

    Returns:
    float or complex: The sum of the series.
    """

    if K == float('inf'):
        # Infinite sum formula for |z| < 1
        if abs(z) >= 1:
            raise ValueError("The series only converges for |z| < 1.")
        return z / (1 - z)
    else:
        # Finite sum formula
        return z * (1 - z**K) / (1 - z)

def polylog(z, q, K):
    """
    Computes the sum: sum_k=1^K (z^k / k^q)
    
    Parameters:
        z (float): The base of the exponential term.
        q (float): The exponent for k.
        K (int): The upper limit of the sum.
    
    Returns:
        float: The computed sum.
    """
    return sum( z**k / (k**q) for k in range(1, K + 1))

def Hi(z, q, K):
    """
    Computes the sum: sum_k=1^K (k^q * z^(2k) / (2k)!)
    
    Parameters:
        z (float): The base of the exponential term.
        q (float): The exponent for k.
        K (int): The upper limit of the sum.
    
    Returns:
        float: The computed sum.
    """
    return sum(k**q * z**(2*k) / math.factorial(2 * k) for k in range(1, K + 1))
```

```python
# Initialize lists
gbs = []
mc = []

# Define a range for n
# n_values = [n * 2 for n in range(1, 15)]
n_values = [n for n in range(4, 30)]

# Populate gbs and mc with some function values
for n in n_values:
    N = n
    K = N**2
    bmin = 0.96/N
    bmax = 1.11/N
    gamma_a = 10/21 * N
    gamma_b = 5/11 * N
    q = -N/2
    
    z_mc = 4 * gamma_a * bmin
    q_mc = q
    v_mc = 1 + np.exp(1/25-1/6) * (1/np.sqrt(np.pi)) * polylog(z_mc, 1/2 - 2*q_mc, K)
    mc.append(v_mc)   # Replace with your desired function for MC
    
    z_gbs = np.sqrt(2 * gamma_b /bmin)
    q_gbsh = q
    h = Hi(z_gbs, q_gbsh, int(N/2))
    q_gbs = -q_gbsh - N/2 + 1/2
    v_gbs = 1 + (1/np.sqrt(np.pi)) * (h+ (2 * np.pi)**((N-1)/2) * N**(q_gbsh - 1/2) * np.exp(N/13) * geometric_series_sum(2*z_gbs/N, int(N)) *  polylog( (z_gbs/N)**N, q_gbs, int(2*N)))
    gbs.append(v_gbs)  # Replace with your desired function for GBS
```

```python
mc = np.array(mc)

def exp_nsq_function(n, a):
    return a * np.exp(n**2)

N = np.array(n_values)
mc = np.array(mc)
# Plot the original data and the cubic fit
plt.figure(figsize=(8, 6))
plt.scatter(N, np.log(mc), color='blue', label=r'$L_{\text{Haf}}^\text{MC}$')
plt.plot(N, N**2*np.log(1.3), color='red', label=r'$a^{N^2}, a = 1.1$')
plt.xlabel('N')
plt.ylabel('log(Values)')
#plt.title(r"Growth Comparison of $U_{\text{Haf}^2}^\text{GBS-I}$ and $N^3$")
plt.legend()
plt.grid(True)
plt.savefig("../rslt/gc-haf-mc-log-n.png", dpi=300, bbox_inches="tight")
plt.show()
```
